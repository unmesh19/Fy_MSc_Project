```python
import ast
```

```python
def fuct(x):
    a = ast.literal_eval(x)
    b = ""
    for i in a:
        b += i + " "
    return b
```

```python
l = ["Plot Kyeword", "Top 5 Casts", "Generes"]
```

```python
for i in l:
    df1[i] = df[i].apply(lambda x : fuct(x))
```

```python
df1.head()
```

```python
df1.info()
```

```python
df1.fillna(" ", inplace = True)
df1.info()
```

```python
df0 = df1["Generes"] + " " + df1["Overview"] + " " + df1["Plot Kyeword"] + " " + df1["Director"] + " " + df1["Top 5 Casts"] + " " + df1["Writer"]
```

```python
df0[0]
```

```python
df0 = df0.apply(lambda x: x.lower())
df0.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vector = CountVectorizer()
```

```python
df_cv = vector.fit_transform(df0)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similar = cosine_similarity(df_cv)
```

```python
movie_list = df["movie title"].to_list()
movie_list
```

```python
from operator import itemgetter as ig
import difflib as dl
```

```python
movie = input("Enter the movie name : ")
best_match = dl.get_close_matches(movie, movie_list)[0]
print(best_match)
index_of_the_movie = df[df["movie title"]==best_match].index[0]
l = list(enumerate(similar[index_of_the_movie]))
match = sorted(l, key = ig(1,0), reverse = True)
final_match = match[1:6]
print("Movies suggested for you : \n")
for i in final_match:
    print(df.iloc[i[0]]["movie title"])
```
